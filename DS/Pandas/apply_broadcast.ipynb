```python
import pandas as pd
import numpy as np
```

```python
def initData(num):
  np.random.seed(42)
  return pd.DataFrame(np.random.randint(1,10,size=[num,num]))
df = initData(10)
df.head(3)
```

```python
df + 10
```

```python
df.applymap(lambda x: x + 10)
```

```python
df.apply(lambda x: x+10)
```

```python
df.applymap(lambda x: print(type(x)))
```

```python
df.apply(lambda x: print(x))
```

```python
def broadcast_func(x):
  if x.name in range(2,5):
    return x + 100
  else:
    return x+10

  
```

```python
df.apply(broadcast_func, axis = 0)
```

```python
def broadcast_func_ext(x):
  if x.name in range(2,5):
    return [val+100 if idx < 5 else val + 50 for idx, val in enumerate(x)]
  else:
    return x+10
```

```python
df.apply(broadcast_func_ext, axis = 0)
```
